```python
pdir = "/glade/campaign/cesm/cesmdata/inputdata/lnd/clm2/paramdata"
defp_fname = "clm5_params.c171117.nc"
Nfixp_fname = "ctsm5.3.041.Nfix_params.v13.c250221_upplim250.nc"

defp = xr.open_dataset(f"{pdir}/{defp_fname}", decode_timedelta=True)
Nfixp = xr.open_dataset(f"{pdir}/{Nfixp_fname}", decode_timedelta=True)
ppe_Nfixp =xr.open_dataset("/glade/u/home/bbuchovecky/projects/cpl_ppe_co2/hist/data/cesm2.1.5/paramfiles/clm5_params.c171117_FUNfix.nc", decode_timedelta=True)

swap = ["kn_nonmyc", "kc_nonmyc"]
```

```python
fig, ax1 = plt.subplots(nrows=3, ncols=1, figsize=(12,9))
ax2 = np.empty_like(ax1)
for i, a in enumerate(ax1):
    ax1[i].tick_params(axis='y', labelcolor='tab:blue')
    ax1[i].set_ylabel("kn_nonmyc", color='tab:blue')

    ax2[i] = ax1[i].twinx()
    ax2[i].tick_params(axis='y', labelcolor='tab:orange')
    ax2[i].set_ylabel("kc_nonmyc", color='tab:orange')

defp["kn_nonmyc"].plot(ax=ax1[0], marker="o", color="tab:blue", alpha=0.5, _labels=False)
defp["kc_nonmyc"].plot(ax=ax2[0], marker="o", color="tab:orange", alpha=0.5, _labels=False)
Nfixp["kn_nonmyc"].plot(ax=ax1[1], marker="o", color="tab:blue", alpha=0.5, _labels=False)
Nfixp["kc_nonmyc"].plot(ax=ax2[1], marker="o", color="tab:orange", alpha=0.5, _labels=False)
(defp["kn_nonmyc"] - Nfixp["kn_nonmyc"]).plot(ax=ax1[2], marker="o", color="tab:blue", alpha=0.5)
(defp["kc_nonmyc"] - Nfixp["kc_nonmyc"]).plot(ax=ax2[2], marker="o", color="tab:orange", alpha=0.5)
```
